```python
"""How to flatten a list in python?

Flattening a list of lists entails converting a 2D list into a 1D list by un-nesting each list item stored in the list of lists - i.e.,
 converting [[1, 2, 3], [4, 5, 6], [7, 8, 9]] into [1, 2, 3, 4, 5, 6, 7, 8, 9]."""

mylist = [[1,2,3],[4,5,6],[7,8,9]]
print("Original List : ",mylist)
print("Flattened List : ",end=" ")
for lst in mylist:
  for val in lst:
    print(val, end= " ")
```

```python
"""We have following information on Employees and their Salary (Salary is in lakhs),
Employee Salary

John 14

Smith 13

Alice 32

Daneil 21

Using above create a dictionary of Employees and their Salary Write a program that asks user for three type of inputs,

print: if user enter print then it should print all Employees with their Salary in this format,

John ==>14

Smith ==>13

Alice ==>32

Daneil ==>21

add: if user input adds then it should further ask for an Employee name to add.
 If Employee already exists in our dataset then it should print that it exists and do nothing.
  If it doesn't then it asks for Salary and add that new Employee/Salary in our dictionary and print it

remove: when user inputs remove it should ask for an Employee to remove.
 If an Employee exists in our dictionary then remove it and print a new dictionary using format shown above in (a). Else print that Employee doesn't exist!

query: on this again ask the user for which Employee he or she wants to query. When a user inputs that Employee it will print the Salary of that Employee."""

di={'John':14,'Smith':13,'Alice':32,'Daniel':21}
inp=input('Enter print/add/remove or query')
if inp=='print':
  for key in di.keys():
    print(f'{key} ==>{di[key]}')
elif inp=='add':
    key=input('Enter employee name : ')
    if key in di:
      print('It exists')
    else:
      sal=input('Enter Salary : ')
      di[key]=sal    
      print(f'{key} ==>{sal}')
elif inp=='remove':
  key=input('Enter employee name : ')
  if key not in di:
    print('Employee doesn\'t exists')
  else:
    val=di[key]
    di.pop(key)
    print(f'{key} ==>{val}')
elif inp=='query':
  key=input("Enter employee name")
  print(di[key])
```
