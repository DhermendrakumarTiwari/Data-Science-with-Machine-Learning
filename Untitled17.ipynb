```python
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.feature_names
```

```python
cancer.target_names
```

```python
df = pd.DataFrame(cancer.data,columns=cancer.feature_names)
df.head()
```

```python
df['target'] = cancer.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df['cancer class']=df.target.apply(lambda x: cancer.target_names[x])
```

```python
df.head()
```

```python
df[50:55]
```

```python
df0 = df[df.target==0].head(50)
df1 = df[df.target==1].head(50)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('mean radius')
plt.ylabel('mean texture')
plt.scatter(df0['mean radius'],df0['mean texture'],color='red',marker='+')
plt.scatter(df1['mean radius'],df1['mean texture'],color='green',marker='*')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','cancer class'],axis='columns')
Y = df.target
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=5)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
```

```python
mymodel=SVC()
```

```python
mymodel.fit(X_train,Y_train)
```

```python
mymodel.score(X_test,Y_test)
```

```python
#testing of our trained model
pred=mymodel.predict(X_test)
#import confusion matrics

from sklearn.metrics import confusion_matrix

cm=np.array(confusion_matrix(Y_test,pred))
cm
```

```python
model_C = SVC(C=1)
model_C.fit(X_train,Y_train)
model_C.score(X_test,Y_test)
```

#Kernel

```python
model_kernel = SVC(kernel='linear')
model_kernel.fit(X_train,Y_train)
model_kernel.score(X_test,Y_test)
```
