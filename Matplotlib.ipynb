#Matplotlib 3-d

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
```

```python
iris = pd.read_csv("Iris.csv")
```

```python
iris
```

```python
for species,irissubset, in iris.groupby('Species'):
  plt.scatter(irissubset['PetalLengthCm'],irissubset['PetalWidthCm'],label=species)
plt.legend()
plt.show()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

for species,irissubset, in iris.groupby('Species'):
  ax.scatter(irissubset['PetalLengthCm'],irissubset['PetalWidthCm'],irissubset['SepalLengthCm'],label=species)
plt.legend()
ax.set_xlabel('PetalLenght')
ax.set_ylabel('PetalWidth')
ax.set_zlabel('SepalWidth')
plt.show()
```

```python
#import matplotlib.pyplot as plt


slices = [12,2,2,30]

activities = ['sleeping','eating','working','playing']

cols = ['c','m','r','b']

plt.pie(slices,

  labels=activities,

  colors=cols,

  startangle=180  )

plt.title('Pie Plot')

plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

def f(t):

    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)

t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(331)  #no of plots, row wise or column wise, 1 2 plot

plt.plot(t1, f(t1), 'bo')

plt.subplot(332)

plt.plot(t2, np.cos(2*np.pi*t2))
plt.subplot(333)  #no of plots, row wise or column wise, 1 2 plot

plt.plot(t1, f(t1), 'bo')



plt.show()
```
