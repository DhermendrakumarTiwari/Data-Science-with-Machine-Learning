```python
class Employee:
  def __init__(self,name,id):
    self.id = id
    self.name = name

  def display(self):
    print("Name %s  \n  ID:  %d"%(self.name,self.id))

emp1 = Employee("Harry",1001)
emp2 = Employee("Smith",1002)

emp1.display()

print(emp1.__doc__)
emp2.display()
```
