# Star Pattern

```python
r=int(input("enter the no of rows"))
for i in range(0,r):
  for i in range(i):
    print("*")
```

```python
r=int(input("enter the no of rows"))
for i in range(0,r):
  for i in range(i):
    print("*",end="")
    
```

```python
r=int(input("enter the no of rows"))
for i in range(0,r):    #r=4  0,4
  for i in range(i):     # i=2  j=0,1
    print("*",end="")
  print()
```
