```python
#Defining string 

str = 'Goeduhub Technologies 12345'

print('str = ', str)

#first character

print('str[0] = ', str[0])

#last character

print('str[-1] = ', str[-1])

#slicing 1nd to 5th character

print('str[1:5] = ', str[1:5])

#slicing 5th to 2nd last character

print('str[5:-2] = ', str[5:-2])

#space cosider as string element 

print('str[8]=',str[8])
```
