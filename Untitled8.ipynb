```
import bs4
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```
url="https://www.mohfw.gov.in/index.html"
response=requests.get(url)
print(response)
```

```
htmlcontent=response.content
#print(htmlcontent)
```

```
soup=BeautifulSoup(htmlcontent,"html.parser")
```

```
print(soup.prettify)
```
