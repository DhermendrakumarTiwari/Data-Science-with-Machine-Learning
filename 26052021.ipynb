#Numpy Vs List Speed

```python
import time
import sys
```

```python
size = 1000000
l1=range(size)
l2=range(size)

n1 = np.arange(size)
n2 = np.arange(size)

#list itemwise sum
start=time.time()
result=[(x+y) for x,y in zip(l1,l2)]
print((time.time()-start)*1000)
#numpy array itemwise sum
start = time.time()
result1 = n1+n2
print((time.time()-start)*1000)
```
