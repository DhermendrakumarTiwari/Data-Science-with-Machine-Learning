#Classification with Multinomial Naive Bayes

```python
from sklearn.datasets import fetch_20newsgroups
```

```python
news = fetch_20newsgroups()
```

```python
print(news.keys())
```

```python
news.target_names
```

```python
cats = ['alt.atheism','soc.religion.christian','comp.graphics','sci.med']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
     'This is the first document.',
     'This document is the second document.',
     'And this is the third one.',
     'Is this the first document?',
      ]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())
```

```python
print(X.toarray())
```

```python
count_vect = CountVectorizer()
news_train = fetch_20newsgroups(subset='train',categories=cats)
```

```python
news_test = fetch_20newsgroups(subset='test',categories=cats)
```

```python
x_train = count_vect.fit_transform(news_train.data)
```

```python
print(x_train.toarray())
```

```python
print(x_train.shape)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer()

x_train_tfidf = tfidf_transformer.fit_transform(x_train)

x_train_tfidf.shape
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB().fit(x_train_tfidf,news_train.target)
```

#Test Model

```python
x_test_df = count_vect.transform(news_test.data)


x_test_tfidf = tfidf_transformer.transform(x_test_df)

pred = model.predict(x_test_tfidf)

pred
```
